```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#x_train is input variable(size in 1000 square feet)
#y_train is the target price(in 1000's of dollars)
x_train = np.array([1.0,2.0])
y_train = np.array([300.0,500.0])
```

```python
# m = number of training examples
m = x_train.shape[0]
```

```python
# use x_i and y_i to denote the ith row in the dataset
i= 0
x_i = x_train[i]
y_i = y_train[i]
```

```python
# plotting the data
plt.scatter(x_train,y_train,marker='x',c='r')
# set the title
plt.title('Housing Prices')
# set y axis label
plt.ylabel('Price (in 1000s of dollars)')
# set x axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# now adjust the w and b parameters for the regression line
w = 100
b =100
```

# we have f_wb is the regression function which gives us the target prediction
# for x^(0), f_wb = w * x[0] + b
# for large data points we will have to create a function and run a for loop inside it

```python
def compute_model_output(x,w,b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      y (ndarray (m,)): target values
    """
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
        
    return f_wb
```

```python
# now call the compute_model_output function and plot the output
```

```python
tmp_f_wb = compute_model_output(x_train,w,b)
# plot model predictions
plt.plot(x_train,tmp_f_wb,c='b',label='Our prediction')
# plot data points
plt.plot(x_train,y_train,marker='x',c='r',label='Actual Values')

# set title
plt.title('Housing Prices')
#set y axis and x axis labels
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# we can see that the values of w and b that we chose are not a good fit for the model
```
